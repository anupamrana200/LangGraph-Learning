```python
from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langchain_openai import ChatOpenAI
from typing import TypedDict, Annotated
from langchain_core.messages import BaseMessage, HumanMessage
from langgraph.checkpoint.memory import MemorySaver
from dotenv import load_dotenv

load_dotenv()
model = ChatOpenAI()
```

```python
class chatState(TypedDict):
  messages: Annotated[list[BaseMessage], add_messages]

  
```

```python
def chatNode(state: chatState):
  messages = state['messages']

  response = model.invoke(messages)

  return {'messages':[response]}
```

```python
checkPointer = MemorySaver()

graph = StateGraph(chatState)

graph.add_node('chatNode', chatNode)

graph.add_edge(START, 'chatNode')
graph.add_edge('chatNode', END)

workflow = graph.compile(checkpointer=checkPointer)
```

```python
initialState = {'messages': [HumanMessage(content='What is the capital of india?')]}

finalResult = workflow.invoke(initialState)

finalResult['messages'][-1].content
```

```python
thread_id = '1'
while True:
  userMessage = input("Type your message here: ")

  print("User: ", userMessage)

  if userMessage.strip().lower() in ['exit', 'quit', 'bye']:
    break


  config = {'configurable':{'thread_id':thread_id}}

  initalMessage = {'messages': [HumanMessage(content=userMessage)]}
  response = workflow.invoke(initalMessage,  config=config)

  print("BOT: ",response['messages'][-1].content)
  # print("\n--- Current State (Full History) ---")
  # for msg in response['messages']:
  #   print(f"{msg.type.upper()}: {msg.content}")
  # print("------------------------------------\n")

```
