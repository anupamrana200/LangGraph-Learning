```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatOpenAI()
```

```python
class llmState(TypedDict):
  question: str
  answer: str
```

```python
def llm_qa(state: llmState) -> llmState:
  question = state['question']

  prompt = f"Answer the following question, here is the question - {question}"

  answer = model.invoke(prompt).content

  state['answer'] = answer

  return state
```

```python
graph = StateGraph(llmState)

graph.add_node('llm_question_answer',llm_qa)

graph.add_edge(START, 'llm_question_answer')
graph.add_edge('llm_question_answer', END)

workflow = graph.compile()
```

```python
initial_state = {'question' : 'What is the distance between moon and earth?'}
final_state = workflow.invoke(initial_state)

print(final_state)
```
